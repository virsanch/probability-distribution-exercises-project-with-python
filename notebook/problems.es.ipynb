# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Ejercicio 1

Escriba las funciones anteriores pero para distribuciones normales y uniformes.

```python
from scipy.stats import binom, norm, uniform  
import numpy as np                            
import matplotlib.pyplot as plt  

```

```python
#Distribucion Binomial

# n=10 intentos, p=0.2 probabilidad de éxito

# ¿Probabilidad de obtener EXACTAMENTE 2 éxitos?
print(binom.pmf(k=2, n=10, p=0.2))

# ¿Probabilidad de obtener 2 éxitos O MENOS?
print(binom.cdf(k=2, n=10, p=0.2))

# ¿Qué valor tiene el cuantil del 90%?
print(binom.ppf(q=0.9, n=10, p=0.2))

# Genera 10 valores aleatorios
print(binom.rvs(n=10, p=0.2, size=10))
```

```python
# Distribución Normal

# media=50, desviación típica=10

# ¿Qué altura tiene la curva en x=55?
print(norm.pdf(55, loc=50, scale=10))

# ¿Probabilidad de obtener 55 O MENOS?
print(norm.cdf(55, loc=50, scale=10))

# ¿Qué valor tiene el cuantil del 90%?
print(norm.ppf(0.9, loc=50, scale=10))

# Genera 10 valores aleatorios
print(norm.rvs(loc=50, scale=10, size=10))
```

```python
# Distribución Uniforme

# min=0, max=10 → todos los valores igual de probables

# ¿Qué densidad tiene el valor 5?
print(uniform.pdf(5, loc=0, scale=10))

# ¿Probabilidad de obtener 5 O MENOS?
print(uniform.cdf(5, loc=0, scale=10))

# ¿Qué valor tiene el cuantil del 90%?
print(uniform.ppf(0.9, loc=0, scale=10))

# Genera 10 valores aleatorios
print(uniform.rvs(loc=0, scale=10, size=10))
```

```python
#Grafica Binomial

fig = plt.figure(figsize=(12, 9))

x_bin = np.arange(0, 11)

plt.subplot(3, 3, 1)
plt.bar(x_bin, binom.pmf(x_bin, n=10, p=0.2), color="#7F77DD")
plt.title("Binomial — PMF")

plt.subplot(3, 3, 2)
plt.step(x_bin, binom.cdf(x_bin, n=10, p=0.2), color="#7F77DD")
plt.title("Binomial — CDF")

plt.subplot(3, 3, 3)
plt.step(np.linspace(0,1,100), binom.ppf(np.linspace(0,1,100), n=10, p=0.2), color="#7F77DD")
plt.title("Binomial — PPF")
plt. show()

```

```python
# Gráfica Normal

fig = plt.figure(figsize=(12, 9))

x_norm = np.linspace(20, 80, 200)
p_vals = np.linspace(0.01, 0.99, 200)

plt.subplot(3, 3, 4)
plt.plot(x_norm, norm.pdf(x_norm, loc=50, scale=10), color="#1D9E75")
plt.title("Normal — PDF")

plt.subplot(3, 3, 5)
plt.plot(x_norm, norm.cdf(x_norm, loc=50, scale=10), color="#1D9E75")
plt.title("Normal — CDF")

plt.subplot(3, 3, 6)
plt.plot(p_vals, norm.ppf(p_vals, loc=50, scale=10), color="#1D9E75")
plt.title("Normal — PPF")
plt. show()
```

```python
# Gráfica Uniforme

fig = plt.figure(figsize=(12, 9))

x_unif = np.linspace(-1, 11, 200)

plt.subplot(3, 3, 7)
plt.plot(x_unif, uniform.pdf(x_unif, loc=0, scale=10), color="#BA7517")
plt.title("Uniforme — PDF")

plt.subplot(3, 3, 8)
plt.plot(x_unif, uniform.cdf(x_unif, loc=0, scale=10), color="#BA7517")
plt.title("Uniforme — CDF")

plt.subplot(3, 3, 9)
plt.plot(p_vals, uniform.ppf(p_vals, loc=0, scale=10), color="#BA7517")
plt.title("Uniforme — PPF")

plt.tight_layout()
plt.show()
```
